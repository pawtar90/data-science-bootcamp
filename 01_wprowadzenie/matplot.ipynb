```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
```

```python
plt.plot(x,x_sin)
plt.plot(x,x_cos)
```

```python
plt.subplot(121)
plt.plot(x,x_sin)
plt.subplot(122)
plt.plot(x,x_cos,'r--')
```

```python
fig = plt.figure(figsize=(7,7))
plt.subplot(131)
plt.plot(x,x_sin)
plt.subplot(132)
plt.plot(x,x_cos,'r--')
plt.subplot(133)
plt.plot(x,x_cos*x_cos,'r--')
```
